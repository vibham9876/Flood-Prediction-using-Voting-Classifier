```python
# Calculate Q1, Q3, and IQR
Q1 = subset_df.quantile(0.25)
Q3 = subset_df.quantile(0.75)
IQR = Q3 - Q1

# Define the range for outlier detection (e.g., 1.5 times IQR)
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Remove outliers from the DataFrame
subset_df_no_outliers = subset_df[(subset_df >= lower_bound) & (subset_df <= upper_bound)]

# Display the DataFrame without outliers
subset_df_no_outliers
```

```python
subset_df_no_outliers.describe()
```

```python
from sklearn.impute import SimpleImputer

# Define imputation strategy (median or mean)
imputation_strategy = 'mean'  # or 'mean'

# Create a SimpleImputer object with the chosen strategy
imputer = SimpleImputer(strategy=imputation_strategy)

# Iterate over each column in the DataFrame
for column in subset_df_no_outliers.columns:
    # Skip the 'FloodProbability' column
    if column == 'FloodProbability':
        continue
    
    # Impute missing values in the current column
    subset_df_no_outliers[column] = imputer.fit_transform(subset_df_no_outliers[[column]])

# Check if there are any remaining missing values after imputation
print(subset_df_no_outliers.isnull().sum())
```

## SVM

```python
from sklearn.svm import SVC
from sklearn.metrics import roc_auc_score
from sklearn.metrics import confusion_matrix
import math

# Define the range of hyperparameters (C values)
C_values = [0.01, 0.1, 1, 10, 100]

train_auc = []
test_auc = []

for C in C_values:
    # Initialize the SVM classifier with the chosen hyperparameter
    clf = SVC(C=C, probability=True)
    
    # Fit the model
    clf.fit(X_train, y_train)
    
    # Predict probabilities for the test set
    prob_cv = clf.predict_proba(X_test)[:, 1]
    
    # Calculate AUC score for test set
    test_auc.append(roc_auc_score(y_test, prob_cv))
    
    # Predict probabilities for the train set
    prob_train = clf.predict_proba(X_train)[:, 1]
    
    # Calculate AUC score for train set
    train_auc.append(roc_auc_score(y_train, prob_train))

# Find the optimal C value
optimal_C = C_values[test_auc.index(max(test_auc))]

# Plot AUC vs hyperparameter
plt.plot([math.log(C) for C in C_values], train_auc, label='AUC train')
plt.plot([math.log(C) for C in C_values], test_auc, label='AUC test')
plt.title('AUC vs hyperparameter')
plt.xlabel('log(C)')
plt.ylabel('AUC')
plt.legend()
plt.show()

print('Optimal C value for which AUC is maximum:', optimal_C)
```

```python
# Train SVM with optimal hyperparameter
from sklearn.metrics import roc_curve
svm = SVC(C=1, probability=True)
svm.fit(X_train, y_train)


# Predict probabilities on test set
pred_test = svm.predict_proba(X_test)[:, 1]

# Calculate ROC curve and AUC on test set
fpr, tpr, thresholds = roc_curve(y_test, pred_test)
roc_auc = roc_auc_score(y_test, pred_test)

# Plot ROC curve
plt.plot(fpr, tpr, label='Test ROC, AUC=' + str(roc_auc))
plt.title('ROC curve')
plt.xlabel('FPR')
plt.ylabel('TPR')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import accuracy_score
pred_test =svm.predict(X_test)
test_accuracy = accuracy_score(y_test, pred_test)
pred_train = svm.predict(X_train)
train_accuracy =accuracy_score(y_train,pred_train)



print("AUC on Test data is " +str(accuracy_score(y_test,pred_test)))
print("AUC on Train data is " +str(accuracy_score(y_train,pred_train)))

print("---------------------------")

# Code for drawing seaborn heatmaps
class_names = ['Flood','No Flood']
df_heatmap = pd.DataFrame(confusion_matrix(y_test, pred_test.round()), index=class_names, columns=class_names )
fig = plt.figure( )
heatmap = sns.heatmap(df_heatmap, annot=True, fmt="d")
```

```python
predicted = svm.predict_proba(X_test[:20])[:,1]
predicted[predicted > 0.5] = 1
predicted[predicted < 0.5] = 0

original = ['No Flood' if x == 1 else 'Flood' for x in y_test[:10]]

pred = []
for i in predicted:
    if int(i) == 1:  # Compare with integer values
        k = 'No Flood'
        pred.append(k)
    else:
        k = 'Flood'
        pred.append(k)

# Creating a data frame
df = pd.DataFrame(list(zip(original, pred)), 
               columns =['original_Classlabel', 'predicted_classlebel'])
df
```

## RandomForestClassifier

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_auc_score
from sklearn.model_selection import GridSearchCV

dept = [1, 5, 10, 50, 100, 500, 1000]
n_estimators =  [20, 40, 60, 80, 100, 120]

param_grid={'n_estimators':n_estimators , 'max_depth':dept}
clf = RandomForestClassifier()
model = GridSearchCV(clf,param_grid,scoring='roc_auc',n_jobs=-1,cv=3)
model.fit(X_train,y_train)
print("optimal n_estimators",model.best_estimator_.n_estimators)
print("optimal max_depth",model.best_estimator_.max_depth)
```

```python
import seaborn as sns
X = []
Y = []
test_auc = []
train_auc = []
for n in n_estimators:
    for d in dept:
        clf = RandomForestClassifier(max_depth = d,n_estimators = n)
        clf.fit(X_train,y_train)
        pred_cv = clf.predict_proba(X_test)[:,1]
        pred_train = clf.predict_proba(X_train)[:,1]
        X.append(n)
        Y.append(d)
        test_auc.append(roc_auc_score(y_test,pred_cv))
        train_auc.append(roc_auc_score(y_train,pred_train))
        
```

```python
optimal_n_estimators = model.best_estimator_.n_estimators
optimal_max_depth = model.best_estimator_.max_depth
```

```python
#training our model for max_depth=10,n_estimators = 120
rf = RandomForestClassifier(max_depth = optimal_max_depth,n_estimators = optimal_n_estimators)
rf.fit(X_train,y_train)


pred_test =rf.predict_proba(X_test)[:,1]
fpr1, tpr1, thresholds1 = roc_curve(y_test, pred_test)
pred_train = rf.predict_proba(X_train)[:,1]
fpr2,tpr2,thresholds2=roc_curve(y_train,pred_train)

#ROC curve
x = plt.subplot( )
x.plot(fpr1, tpr1, label='Test ROC ,auc='+str(roc_auc_score(y_test,pred_test)))
x.plot(fpr2, tpr2, label='Train ROC ,auc='+str(roc_auc_score(y_train,pred_train)))
plt.title('ROC')
plt.xlabel('FPR')
plt.ylabel('TPR')
x.legend()
plt.show()

print("AUC on Test data is " +str(roc_auc_score(y_test,pred_test)))
print("AUC on Train data is " +str(roc_auc_score(y_train,pred_train)))

print("---------------------------")

# Code for drawing seaborn heatmaps
class_names = ['No Flood','Flood']
df_heatmap = pd.DataFrame(confusion_matrix(y_test, pred_test.round()), index=class_names, columns=class_names )
fig = plt.figure( )
heatmap = sns.heatmap(df_heatmap, annot=True, fmt="d")
```
